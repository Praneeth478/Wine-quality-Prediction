# Data Preprocessing

```python
def wisker(col):
    q1,q3=p.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
for i in Dataset.iloc[:,0:11].select_dtypes(include="number").columns:
    lw,uw=wisker(Dataset[i])
    Dataset[i]=p.where(Dataset[i]<lw,lw,Dataset[i])
    Dataset[i]=p.where(Dataset[i]>uw,uw,Dataset[i])
```
